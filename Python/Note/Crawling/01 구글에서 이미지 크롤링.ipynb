```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd

import time

from urllib.request import urlopen
import urllib.request
```

```python
# Chrome Browser와ㅏ Chrome Driver Version 확인하기
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.get('https://www.google.com/')

#검색창에 원하는 검색어 입력.
element = driver.find_element(By.CLASS_NAME, 'gLFyf')
element.send_keys('saxophone')  # 여기에서 검색값 변경
time.sleep(0.5)
element.submit()
time.sleep(0.5)

# 이미지 클릭
xpath = '//*[@id="hdtb-msb"]/div[1]/div/div[2]'
gu_list_raw = driver.find_element(By.XPATH, xpath).click()

SCROLL_PAUSE_TIME = 1

last_height = driver.execute_script("return document.body.scrollHeight") #현재 스크롤 위치

new_height = 0

while True :

    driver.execute_script("window.scrollBy(0,document.body.scrollHeight)") 
    # 첫 번째 인수에 있는 0 값은 가로 방향으로 스크롤을 하지 않도록
    # 전체 웹페이지의 높이(document.body.scrollHeight)에 해당하는 픽셀 수만큼 세로 방향으로 스크롤

    time.sleep(SCROLL_PAUSE_TIME)

    new_height = driver.execute_script("return document.body.scrollHeight")

    if new_height == last_height:
        try:
            driver.find_element(By.CLASS_NAME, "mye4qd").click()
        except:
            break
    last_height = new_height
```

```python
img_elements = driver.find_elements(By.CSS_SELECTOR,".rg_i") # 모든 이미지의 css값 가져오기
len(img_elements)
```

```python
#이미지 저장
for idx, img in enumerate(img_elements) :

    print("색소폰 : %s/%s"%(idx+1,len(img_elements)))

    try :

        img.click()

        time.sleep(SCROLL_PAUSE_TIME)

        img_element = driver.find_element(By.XPATH,'//*[@id="Sva75c"]/div[2]/div/div[2]/div[2]/div[2]/c-wiz/div/div[1]/div[2]/div[2]/div/a/img').get_attribute('src')
        urllib.request.urlretrieve(img_element, "../../Data/plus_data/sexophone/sexophone%04d.jpg"%idx) #파일 저장 경로 수정
        
    except :
        print('error')
        pass

driver.close()
```
