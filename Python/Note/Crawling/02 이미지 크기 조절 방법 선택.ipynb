```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Cubic convolution interpolation 기법으로 이미지 크기 조정
from scipy import misc
from scipy import ndimage

# 이미지 파일을 읽어옵니다.
img = misc.ascent()

# 이미지 크기를 줄이기 위한 크기를 계산합니다.
height, width = img.shape
print(img.shape) # (100,100)으로 조절해주기 위한 shape 확인
# Cubic Convolution Interpolation을 사용하여 이미지 크기를 변경합니다.
newCubic_img = ndimage.interpolation.zoom(img, zoom=(100/512, 100/512), order=3)
```

```python

import matplotlib.pyplot as plt
from PIL import Image

# 이미지 파일을 읽어옵니다.
img = misc.ascent()

# 이미지 크기를 100x100으로 변경합니다.
LANCZOS_img = Image.fromarray(img).resize((100,100), Image.LANCZOS)
```

```python
# 결과 이미지를 출력합니다.
plt.imshow(LANCZOS_img, cmap='gray') # LANCZOS 기법으로 (100,100)
plt.show()

plt.imshow(newCubic_img, cmap='gray') # Cubic convolution interpolation 기법으로 (100,100)
plt.show()
```

### Cubic convolution interpolation 기법으로 이미지 크기 조정하는게 특성을 좀 더 살리면서 크기조절이 되는 것 같다.
