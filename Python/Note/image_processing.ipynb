```python
import glob
from PIL import Image

filelists=glob.glob('../Data/saved/acoustic/*.jpg')
filelists[:5]
```

```python
import glob
from PIL import Image
import numpy as np

folderlist=['acoustic','electric','saxophone']
fileCount=0

for folders in folderlist:
    for filename in glob.glob(f'../Data/saved/{folders}/*.jpg'):
        image=Image.open(filename)
        image=image.convert('L')
        imgArray=np.array(image)
        imgc=255-imgArray
        img=Image.fromarray(imgc)
        for angle in range(-90,90,15):
            img11=img.rotate(angle)
            img11.save(f'../Data/converted/{folders}/{folders}{fileCount}_angle_{angle}.jpg','JPEG')
            img111=img11.transpose(Image.FLIP_LEFT_RIGHT)
            img11.save(f'../Data/converted/{folders}/{folders}{fileCount}_angle_{angle}_flipped.jpg','JPEG')
            fileCount+=1
    fileCount=0
```

```python
import numpy as np

imglist=[]

for img in sorted(filelists):
    image=Image.open(img)
    image=image.convert('L')
    imgArray=np.array(image)
    imglist.append(imgArray)

imgarrays=np.array(imglist)
```

```python
imgConverted=[]

for i in imgarrays:
    imgc=255-i
    imgConverted.append(imgc)

imgConverted=np.array(imgConverted)
```

```python
import matplotlib.pyplot as plt

plt.imshow(imgConverted[0],cmap='gray')
plt.show()
```

```python
k=0
for j in imgConverted:
    img=Image.fromarray(j)
    img.save('../Data/blueconverted/electric_guitar_%d.jpg'%k)
    k+=1
```

```python
fileCount=0
blue=glob.glob('../Data/blueconverted/*.jpg')

for filename in blue:
    for angle in range(-90,90,15):
        img=Image.open(filename)
        img11=img.rotate(angle)
        img11.save('../Data/electric_guitar/blue%d_angle_%d.jpg'%(fileCount,angle),'JPEG')
        img111=img11.transpose(Image.FLIP_LEFT_RIGHT)
        img11.save('../Data/electric_guitar/blue%d_angle_%d_flipped.jpg'%(fileCount,angle),'JPEG')
        fileCount+=1
```
